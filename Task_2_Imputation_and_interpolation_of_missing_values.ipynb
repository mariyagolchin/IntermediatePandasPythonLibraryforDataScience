```
# Handeling Missing Values

# Mean Impute Missing Value
# Interpolate Missing value  ===>>> if data has linear relationship use Interpolation->  30 nul 40   -> 30 , 40 is linear growth

import numpy as np
import pandas as pd

df=pd.read_csv("https://raw.githubusercontent.com/VinitaSilaparasetty/Coursera-Intermediate-Pandas/master/What_does_aid_to_Africa_finance_1.csv ")

df.head(3) #Ensure data has loaded correctly.
```

## Mean Impute Missing Value

```

# Mean Impute Missing Value

# pd.isnull(df).sum()
df.isnull().sum()
```

```
# popn has 4 missing value

df['popn']
```

```
df['popn'].mean()
```

```
df.isnull()
```

```
df['popn'].fillna(df['popn'].mean(),inplace=True)  #replace non in popon with mean in popn
```

```
df.isnull().sum()
```

## Interpolate Missing value

```
df1=pd.read_csv("https://raw.githubusercontent.com/VinitaSilaparasetty/Coursera-Intermediate-Pandas/master/What_does_aid_to_Africa_finance_1.csv ")

df1.head(3) #Ensure data has loaded correctly.
df1.isnull().sum()
```

```
df1['popn'].fillna(df1['popn'].interpolate(),inplace=True)
df1.isnull().sum()
```
